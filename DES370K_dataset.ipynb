```python
import matplotlib.font_manager
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 22}

matplotlib.rc('font', **font)
```

```python
def ternary(sapt, t, dsname, title='', labeled=True, view=True,
            saveas='', relpath=False, graphicsformat=['pdf']):
    """Takes array of arrays *sapt* in form [elst, indc, disp] and builds formatted
    two-triangle ternary diagrams. Either fully-readable or dotsonly depending
    on *labeled*. Saves in formats *graphicsformat*.

    """
    import hashlib
    import numpy as np
    import matplotlib.pyplot as plt
    import matplotlib as mpl
    from matplotlib.path import Path
    import matplotlib.patches as patches
    import warnings
    # Suppress all warnings
    warnings.filterwarnings("ignore")
    # Reset warning behavior to default (optional)
    warnings.filterwarnings("default")
    unique_group_ids = data['group_id'].unique()

    # initialize plot
    fig, ax = plt.subplots(figsize=(6, 3.6))
    plt.xlim([-0.75, 1.25])
    plt.ylim([-0.18, 1.02])
    plt.xticks([])
    plt.yticks([])
    ax.set_aspect('equal')

    if labeled:
        # form and color ternary triangles
        codes = [Path.MOVETO, Path.LINETO, Path.LINETO, Path.CLOSEPOLY]
        pathPos = Path([(0., 0.), (1., 0.), (0.5, 0.866), (0., 0.)], codes)
        pathNeg = Path([(0., 0.), (-0.5, 0.866), (0.5, 0.866), (0., 0.)], codes)
        ax.add_patch(patches.PathPatch(pathPos, facecolor='white', lw=2))
        ax.add_patch(patches.PathPatch(pathNeg, facecolor='#fff5ee', lw=2))

        # form and color HB/MX/DD dividing lines
        ax.plot([0.667, 0.5], [0., 0.866], color='#eeb4b4', lw=0.5)
        ax.plot([-0.333, 0.5], [0.577, 0.866], color='#eeb4b4', lw=0.5)
        ax.plot([0.333, 0.5], [0., 0.866], color='#7ec0ee', lw=0.5)
        ax.plot([-0.167, 0.5], [0.289, 0.866], color='#7ec0ee', lw=0.5)
        ax.plot([-0.167, 0.5], [0.289, 0.866], color='#7ec0ee', lw=0.5)


        # label corners
        ax.text(1.0, -0.15, u'Elst (\u2212)',
            verticalalignment='bottom', horizontalalignment='center',
            family='Serif', weight='bold', fontsize=14)
        ax.text(0.5, 0.9, u'Indx (\u2212)',
            verticalalignment='bottom', horizontalalignment='center',
            family='Serif', weight='bold', fontsize=14)
        ax.text(0.0, -0.15, u'Dispx (\u2212)',
            verticalalignment='bottom', horizontalalignment='center',
            family='Serif', weight='bold', fontsize=14)
        ax.text(-0.5, 0.9, u'Elst (+)',
            verticalalignment='bottom', horizontalalignment='center',
            family='Serif', weight='bold', fontsize=14)
        ax.text(0.0, 1.0, dsname,
            verticalalignment='bottom', horizontalalignment='center',
            family='Serif', weight='normal', fontsize=18)

    xvals = []
    yvals = []
    cvals = []
    for sys in sapt:
        [elst, indc, disp] = sys
       # print(sys)
        # calc ternary posn and color
        Ftop = abs(indc) / (abs(elst) + abs(indc) + abs(disp))
        Fright = abs(elst) / (abs(elst) + abs(indc) + abs(disp))
        xdot = 0.5 * Ftop + Fright
        ydot = 0.866 * Ftop
        cdot = 0.5 + (xdot - 0.5) / (1. - Ftop)
        if elst > 0.:
            xdot = 0.5 * (Ftop - Fright)
            ydot = 0.866 * (Ftop + Fright)
                    #print elst, indc, disp, '', xdot, ydot, cdot

        xvals.append(xdot)
        yvals.append(ydot)
        cvals.append(cdot)

    sc = ax.scatter(xvals, yvals, c=cvals, s=1, marker=(5,2), \
        cmap=mpl.cm.jet, edgecolor='none', vmin=0, vmax=1, zorder=10)

    # remove figure outline
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)

    # save and show
   # pltuid = title + '_' + ('lbld' if labeled else 'bare') + '_' + hashlib.sha1(title + repr(sapt)).hexdigest()
   # pltfile = expand_saveas(saveas, pltuid, def_prefix='tern_', relpath=relpath)
   # files_saved = {}
   # for ext in graphicsformat:
   #     savefile = pltfile + '.' + ext.lower()
   #     plt.savefig(savefile, transparent=True, format=ext, bbox_inches='tight',
   #                 frameon=False, dpi=450, edgecolor='none', pad_inches=0.0)
   #     files_saved[ext.lower()] = savefile
   # if view:
    plt.savefig(t)
    plt.show()
    plt.close()
    return
```

```python
import warnings
# Suppress all warnings
warnings.filterwarnings("ignore")
# Reset warning behavior to default (optional)
warnings.filterwarnings("default")
unique_group_ids = data['group_id'].unique()
#print(unique_group_ids)
#filtered_df = data[data['group_id']==unique_group_ids[0]]

# Now calling value_counts on the 'group_id' column of the filtered dataframe
group_id_counts = filtered_df[['group_id','cbs_CCSD(T)_all','sapt_es','sapt_ind','sapt_disp']]
min_value = group_id_counts['cbs_CCSD(T)_all'].min()
aa=group_id_counts[group_id_counts['cbs_CCSD(T)_all']==min_value].iloc[0]
sapt=[]
for i in range(len(unique_group_ids)):
#for i in range(100):
  filtered_df = data[data['group_id']==unique_group_ids[i]]
  group_id_counts = filtered_df[['group_id','cbs_CCSD(T)_all','sapt_es','sapt_ind','sapt_disp']]
  min_value = group_id_counts['cbs_CCSD(T)_all'].min()
  aa=group_id_counts[group_id_counts['cbs_CCSD(T)_all']==min_value].iloc[0]
  elec=aa[2]
  ind=aa[3]
  disp=aa[4]
  sapt.append([elec,ind,disp])
  #elec=data['sapt_es'][i]
  #ind=data['sapt_ind'][i]
  #disp=data['sapt_disp'][i]
  #sapt.append([elec,ind,disp])
dsname='DES370K'
t="DES370K.pdf"
ternary(sapt, t, dsname, title='', labeled=True, view=True, saveas=None, relpath=False, graphicsformat=['pdf'])
```
